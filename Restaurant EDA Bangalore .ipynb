# Importing libraries 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
plt.style.use('dark_background')
```

# reading csv

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

Dropping Unwanted Columns

```python
df= df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

# cleaning rate columns

```python
df['rate'].unique()
```

```python
def changeRate (value):
    if(value=='NEW' or value=='-'):
        return np.nan
    
    else:
        value = str(value).split('/')
        value = value[0]
        
        return float(value)
    
df['rate'] = df['rate'].apply(changeRate);
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

# filling null values in Rate with mean

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'Type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

# Deleting extra city columns

```python
df = df.drop(['listed_in(city)'],axis=1)
df.head()
```

# flitering cost2plates

```python
df['cost2plates'].unique()
```

```python
def checkCost(value):
    value = str(value)
    if ',' in value :
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost2plates']= df['cost2plates'].apply(checkCost)
df.head()
```

# Filtering Rest Type

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending= False )
rest_type
```

```python
rest_type_less_than_1000 = rest_type[rest_type<1000]
rest_type_less_than_1000
```

```python
#making this as OTHERS
```

```python
def changeRestType(value):
    if(value in rest_type_less_than_1000):
        return 'Others'
    
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(changeRestType)
df['rest_type'].value_counts()
```

# Filtering Location columns

```python
location = df['location'].value_counts(ascending = False)
location_less_than_300 = location[location<300]

```

```python
def changeLocation(value):
    if (value in location_less_than_300):
        return 'Others'
    else:
        return value
    
df['location'] = df['location'].apply(changeLocation)
```

```python
df['location'].value_counts().head()
```

# Filtering Cusine columns

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines_less_than_100 = cuisines[cuisines<100]

def changeCuisines(value):
    if(value in cuisines_less_than_100):
        return 'Others'
    
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(changeCuisines)
df['cuisines'].value_counts()
```

```python
df['Type'].value_counts()
```
